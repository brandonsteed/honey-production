```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
% matplotlib inline

df = pd.read_csv('honeyproduction.csv')
```

```python
year1998 = df.loc[(df['year']==1998)]
year2012 = df.loc[(df['year']==2012)]
```

```python
sns.distplot(df['yieldpercol'], kde=False, rug=True, norm_hist=False)
plt.plot()
```

This histogram with rug plot shows the distribution of honey yields per colony.  Histograms obscure individual data points since they are graphed as a range.  The rug plot mitigates the issue of not seeing individual data points somewhat, but the data points are still kind of difficult to see.

```python
sns.kdeplot(df['yieldpercol'], shade=True, color='orange')
plt.plot()
```

The density curve looks very nice, but individual data is completely hidden.

```python
sns.boxplot(x='year', y='yieldpercol', data=df, palette='PRGn')
plt.plot()
```

The box plots provide a nice visualization for median and quartiles, as well as the range and outliers.  Again, individual data points are hidden, but you can quickly see what is happening with the median and the range.

```python
sns.set_style("whitegrid")
sns.distplot(df['yieldpercol'], color="m")
plt.plot()
```

This is a combination of the density curve and the histogram.  It does show how the density curve doesn't exactly follow the histogram (look at 54-60).  Again, individual data is hidden.

```python
sns.jointplot(x='priceperlb', y='numcol', data=df)
plt.plot()
```

The scatter plot shows each data point.  It does get crowded at the bottom. making it hard to see clearly.

```python
with sns.axes_style("white"):
    sns.jointplot(x='priceperlb', y='totalprod', data=df, kind="hex", color="k")
```

This "hex plot" seems to be somewhat fuzzy and hard on the eyes.  It does have the benefit of showing darker gray to show density instead of having crowded points.

```python
sns.jointplot(x='priceperlb', y='totalprod', data=df, kind="kde")
plt.plot()
```

This joint density estimate is much clearer than the scatter plot, but individual data is not shown.

```python
southeast = df[df['state'].isin(['GA','NC','SC','TN','MS','AL','LA'])]
```

```python
sns.set_style("whitegrid")
sns.factorplot(x='state', y='totalprod', data=southeast, 
                   size=8, kind="bar", palette="muted")
plt.plot()
```

The bar plot provides a visualization for size and for error.  Some people find bar plots difficult to read.

```python
sns.stripplot(x='state', y='totalprod',
              data=southeast, dodge=True, jitter=True,
              size=8, alpha=.25, zorder=1)
plt.plot
```

This strip plot provides a good visualization for the range of each category.  Individual data points are shown.

```python
sns.tsplot(data=southeast, time='year', unit='state',
           condition='state', value='totalprod')
plt.ylim(0,6000000)
plt.xlim(1998, 2012)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

The line plot provides a good visualization for changes in the total production for each state.

```python
sns.boxplot(x='totalprod', y='state', data=southeast,
            whis=np.inf, palette="muted")
```

The box plots show median, quartiles, and range for total production of honey.  Some of them are difficult to see.

```python
sns.set(style="whitegrid")
sns.barplot(x='totalprod', y='state', data=southeast, palette="muted")
plt.plot()
```

The bar plot shows the mean of total production along with range.  Individual data is hidden.

```python
sns.factorplot(x='state', y='totalprod', data=southeast,
                   size=6, kind="point", palette="muted",
                   ci=95,dodge=True,join=False)
```

The pointplot is probably a better visualization than the bar plot.  Mean and range is shown in a more compact manner.

```python
sns.set(style='darkgrid')
sns.pairplot(southeast[['numcol', 'yieldpercol', 'totalprod', 'priceperlb', 'state']], hue='state', palette=sns.hls_palette(n_colors=7, l=.53))
plt.plot()
```

This shows the relationship between the different variables within the different states, provided a large amount of information.
